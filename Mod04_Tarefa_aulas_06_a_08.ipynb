####  2) Melhorando a interação com o API
Observe dois URLs de consultas diferentes, por exemplo o URL utilizado em aula, e o URL feito no exercício anterior. Compare-os e observe as diferenças.

1. Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado (```'Estado'```ou ```Município```), e que devolve os dados da consulta no formato *DataFrame*.
2. Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com *status* em 'Arquivado por decurso de prazo' estão registradas?
3. Qual é o município da Bahia com mais solicitações deferidas?
4. Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'

````python
#1) Seu código aqui
# URLs de consulta
url_aula = 'http://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl'
url_exercicio_anterior = 'http://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf=SP&tipo_interessado=Estado'

def consultar_dados_api(uf, tipo_interessado):
    # URL base da API
    url_base = 'http://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl'

    # Construindo o URL de consulta com base nos parâmetros fornecidos
    url_consulta = url_base
    if uf:
        url_consulta += f'?uf={uf}'
    if tipo_interessado:
        url_consulta += f'&tipo_interessado={tipo_interessado}'

    try:
        # Fazendo a requisição GET
        response = requests.get(url_consulta)

        # Verificando se a requisição foi bem-sucedida
        if response.status_code == 200:
            # Convertendo os dados JSON para DataFrame
            df = pd.DataFrame(response.json()['items'])
            return df
        else:
            print("Erro ao fazer a requisição. Código de status:", response.status_code)
            return None
    except Exception as e:
        print("Erro:", e)
        return None
````

````python
# Exemplo de uso da função com o URL da aula
uf_aula = 'SP'
tipo_interessado_aula = ''
dados_aula = consultar_dados_api(uf_aula, tipo_interessado_aula)
if dados_aula is not None:
    print("Dados da consulta da aula:")
    print(dados_aula.head())  # Exibindo os primeiros registros do DataFrame
else:
    print("Não foi possível obter os dados da consulta da aula.")
````

````python
# 2) Seu código aqui
def contar_solicitacoes_arquivadas_por_estado(url1, url2, uf):
    # Lista para armazenar o número de solicitações de cada URL
    num_solicitacoes = []

    # URLs
    urls = [url1, url2]

    # Loop sobre as URLs
    for url in urls:
        # Adiciona os parâmetros à URL
        url += f'?uf={uf}&status=Arquivado por decurso de prazo'

        # Faz a solicitação HTTP
        response = requests.get(url)

        # Verifica se a solicitação foi bem-sucedida
        if response.status_code == 200:
            # Transforma os dados da resposta em um DataFrame
            df = pd.DataFrame(response.json()['items'])

            # Adiciona o número de solicitações à lista
            num_solicitacoes.append(len(df))
        else:
            print(f"Erro na requisição: {response.status_code}")
            return None

    # Retorna a soma dos números de solicitações
    return sum(num_solicitacoes)

# URLs
url_aula = 'http://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl'
url_exercicio_anterior = 'http://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf=SP&tipo_interessado=Estado'

# Conta o número de solicitações para Minas Gerais com status 'Arquivado por decurso de prazo' para ambas as URLs
num_solicitacoes_total = contar_solicitacoes_arquivadas_por_estado(url_aula, url_exercicio_anterior, 'MG')
num_solicitacoes_total
````

````python
# 3) Seu código aqui
num_solicitacoes_ba = contar_solicitacoes_arquivadas_por_estado(url_aula, url_exercicio_anterior,'BA')
num_solicitacoes_ba
````

````python
# 4) Seu código aqui
# Exemplo de uso da função com o URL da aula
uf_aula = 'BA'
tipo_interessado_aula = ''
dados_aula = consultar_dados_api(uf_aula, tipo_interessado_aula)
dados_aula
# Salvar os dados em um arquivo .csv
if dados_aula is not None:
    dados_aula.to_csv('solicitacoes_bahia_estado.csv', index=False)
    print("Arquivo CSV salvo com sucesso.")
else:
    print("Não foi possível salvar o arquivo CSV.")
````

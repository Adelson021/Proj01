```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('sinasc_RO_2019.csv', sep=',')
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
plt.figure(figsize=(6, 4))
df.boxplot(column='IDADEMAE')
plt.title('Boxplot da Idade da Mãe')
plt.ylabel('Idade')
plt.show()
```

```python
df[['IDADEMAE']].boxplot()
```

####  2) Faça um boxplot do peso do bebê

```python
plt.figure(figsize=(6, 4))
df.boxplot(column='PESO')
plt.title('Boxplot do Peso do Bebê')
plt.ylabel('Peso (kg)')
plt.show()
```

```python
df[['PESO']].boxplot()
```

####  3) Faça um histograma da idade da mãe

```python
plt.figure(figsize=(10, 6))
df['IDADEMAE'].plot(kind='hist', bins=20, edgecolor='black')
plt.title('Histograma da Idade da Mãe')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
```

```python
df['IDADEMAE'].plot.hist(bins = 20, edgecolor='black')
```

#### 4) Faça um histograma do peso do bebê

```python
plt.figure(figsize=(10, 6))
df['PESO'].plot(kind='hist', bins=20, edgecolor='black')
plt.title('Histograma do peso do bebê')
plt.xlabel('Peso')
plt.ylabel('Frequência')
plt.show()
```

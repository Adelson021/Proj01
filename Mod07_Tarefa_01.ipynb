````python
# Criar um DataFrame para armazenar os metadados
metadados = pd.DataFrame({
    'nome_variavel': df.columns,
    'tipo_dados': df.dtypes
})

# Inicializar uma lista vazia para armazenar a quantidade de categorias
qtd_categorias = []

# Loop ao longo dos nomes das variáveis para contar o número de categorias
for col in df.columns:
    qtd_categorias.append(df[col].nunique())

# Adicionar a lista qtd_categorias ao DataFrame de metadados
metadados['qtd_categorias'] = qtd_categorias

# Exibir o DataFrame de metadados
print(metadados)
````

#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
#Seleção de variáveis qualitativas
qualitativas = df.select_dtypes(include=["object", "category"]).columns
qualitativas
````

````python
#Criação de variáveis dummy
df_dummies = pd.get_dummies(df, columns=qualitativas, drop_first=True)
df_dummies
````

````python
#Filtragem para manter apenas variáveis metálicas e booleanas
variaveis_numericas = df_dummies.select_dtypes(include=["number", "bool"])
variaveis_numericas
````

````python
print("Primeiras linhas do dataframe com variáveis dummy:")
print(df_dummies.head())

print("\nPrimeiras linhas do dataframe final adequado para scikit-learn:")
print(variaveis_numericas.head())
````

#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
cross_tab_email = pd.crosstab(df["possui_email"], df["mau"], normalize="index").reindex(columns=[0, 1], fill_value=0)
cross_tab_veiculo = pd.crosstab(df["posse_de_veiculo"], df["mau"], normalize="index").reindex(columns=[0, 1], fill_value=0)

print("Tabela cruzada entre possui_email e mau:")
print(cross_tab_email)
print("\nTabela cruzada entre posse_de_veiculo e mau:")
print(cross_tab_veiculo)
````

````python
diff_email = cross_tab_email[1].max() - cross_tab_email[1].min()
diff_veiculo = cross_tab_veiculo[1].max() - cross_tab_veiculo[1].min()
print("\nDiferença na proporção de 'mau' = 1 para possui_email:", diff_email)
print("Diferença na proporção de 'mau' = 1 para posse_de_veiculo:", diff_veiculo)
````

````python
# Determinar qual variável é mais poderosa
if diff_email > diff_veiculo:
    print("\nA variável 'possui_email' é mais poderosa para prever a probabilidade de 'mau' = 1.")
else:
    print("\nA variável 'posse_de_veiculo' é mais poderosa para prever a probabilidade de 'mau' = 1.")
````

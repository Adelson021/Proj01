```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('sinasc_RO_2019.csv', sep=',')
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
variaveis = ['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']

# Criar um gráfico de barras para cada variável
for var in variaveis:
    counts = df[var].value_counts()
    plt.figure(figsize=(6, 4))
    counts.plot(kind='bar', color='skyblue')
    plt.title(f'{var}')
    plt.xlabel(var)
    plt.ylabel('Contagem')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
apg5_media_por_consultas = df.groupby('CONSULTAS')['APGAR5'].mean().reset_index()
apg5_media_por_consultas
```

```python
apg5_media_por_consultas = apg5_media_por_consultas.sort_values(by='APGAR5', ascending=False)
apg5_media_por_consultas
```

```python
plt.figure(figsize=(10, 6))
plt.bar(apg5_media_por_consultas['CONSULTAS'], apg5_media_por_consultas['APGAR5'], color='skyblue')
plt.title('Média do APGAR5 por CONSULTAS')
plt.xlabel('CONSULTAS')
plt.ylabel('Média do APGAR5')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Gráfico 1: Frequência de SEXO
sexo_counts = df['SEXO'].value_counts()
axs[0, 0].bar(sexo_counts.index, sexo_counts.values, color='skyblue')
axs[0, 0].set_title('Frequência de SEXO')
axs[0, 0].set_xlabel('SEXO')
axs[0, 0].set_ylabel('Frequência')
axs[0, 0].tick_params(axis='x', rotation=0)  # Rotaciona as etiquetas em 30 graus

# Gráfico 2: Frequência de GESTACAO
gestacao_counts = df['GESTACAO'].value_counts()
axs[0, 1].bar(gestacao_counts.index, gestacao_counts.values, color='skyblue')
axs[0, 1].set_title('Frequência de GESTACAO')
axs[0, 1].set_xlabel('GESTACAO')
axs[0, 1].set_ylabel('Frequência')
axs[0, 1].tick_params(axis='x', rotation=30)  # Rotaciona as etiquetas em 30 graus

# Gráfico 3: Frequência de ESTCIVMAE
estcivmae_counts = df['ESTCIVMAE'].value_counts()
axs[1, 0].bar(estcivmae_counts.index, estcivmae_counts.values, color='skyblue')
axs[1, 0].set_title('Frequência de ESTCIVMAE')
axs[1, 0].set_xlabel('ESTCIVMAE')
axs[1, 0].set_ylabel('Frequência')
axs[1, 0].tick_params(axis='x', rotation=0)  # Rotaciona as etiquetas em 30 graus

# Gráfico 4: Frequência de CONSULTAS
consultas_counts = df['CONSULTAS'].value_counts()
axs[1, 1].bar(consultas_counts.index, consultas_counts.values, color='skyblue')
axs[1, 1].set_title('Frequência de CONSULTAS')
axs[1, 1].set_xlabel('CONSULTAS')
axs[1, 1].set_ylabel('Frequência')
axs[1, 1].tick_params(axis='x', rotation=0)  # Rotaciona as etiquetas em 30 graus

# Ajustar layout e exibir os gráficos
plt.tight_layout()
plt.show()
```
